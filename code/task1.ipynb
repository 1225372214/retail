```python
from pandas import*
import pandas as pd
from pandas import Series,DataFrame,np
from numpy import nan as NA
import datetime
P=open(r'D:/实习/附件1.csv')
df=pd.read_csv(P,encoding='utf-8')
p=DataFrame(df)
data_1=p.dropna()#数据过滤
data_11=data_1.drop_duplicates()#去除重复值
print (data_11.describe())#检查是否存异常数据金额
print(data_11.shape)#计算出表格的行列数值
print(data_11.columns)#展示表格列的内容

#转换表格内时间类型
 #del DataFrame.data_11[:,-1]
print("使用read_csv读取的订单信息表的长度为：", len(data_11))
data_11["支付时间"] = pd.to_datetime(data_11["支付时间"])
print('进行转换后表的类型为：\n',  data_11["支付时间"].dtypes)
#提取某月数据
year = [i.year for i in data_11["支付时间"]]## 提取年份信息
month = [i.month for i in data_11["支付时间"]]## 提取月份信息
day = [i.day for i in  data_11["支付时间"]]## 提取日期信息
week = [i.week for i in  data_11["支付时间"]]## 提取周信息
weekday = [i.weekday() for i in  data_11["支付时间"]]##提取星期信息
weekname = [i.weekday_name for i in data_11["支付时间"]]## 提取星期名称信息
timemin = data_11["支付时间"].min()
timemax =data_11["支付时间"].max()
print('订单最早的时间为：',timemin)
print('订单最晚的时间为：',timemax)

data_11 = data_11.set_index("支付时间")
data_11.to_csv('D:/DATA泰迪/new附件1.csv',encoding='utf-8')#存入文件名为new附件1.csv的表
month5_data=data_11.loc["2017-05"]
print(month5_data)
month5_data.to_csv('D:/DATA泰迪/month5 data.csv',encoding='utf-8')#将提取的月份数据存入csv文件中
month6_data=data_11.loc["2017-06"]
month6_data.to_csv('D:/DATA泰迪/month6 data.csv',encoding='utf-8')
```
